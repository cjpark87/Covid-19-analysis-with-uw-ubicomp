# Second we aggregate the same county's data on the same date

At this point we could upload the 8GB to Google Drive and use Google Colab, but uploading is so slow, so I decide to reduce the dataset further more on my local computer. Namely, I want to aggregate the same county on the same date data.

```python
# first let's use a small data to see if our idea works.
data = pd.read_csv("social_dist_all_reduced.csv", nrows = 50000)
```

```python
data.head()
```

```python
# convert device_count column to int (it is string)
data['device_count'] = pd.to_numeric(data['device_count'], errors='coerce') 
```

```python
# convert completely_home_device_count column to int
data['completely_home_device_count'] = pd.to_numeric(data['completely_home_device_count'], errors='coerce')
```

```python
# groupby data in a single swoop (50000 data at once)
data.groupby(['date_range_start', 'state', 'state_code',\
             'cnamelong', 'county_code']).agg(device_count = ('device_count', 'sum'),\
                                             completely_home_device_count = ('completely_home_device_count', 'sum')).reset_index()
```

```python
# inject the header into the dataframe
startingDataFrame = pd.read_csv("social_dist_all_reduced.csv", nrows=0)
startingDataFrame
```

```python
# group by date in chunks (1000 data a chunk)
for chunk in pd.read_csv("social_dist_all_reduced.csv", nrows=50000, chunksize = 1000):
    startingDataFrame = startingDataFrame.append(chunk, ignore_index = True)
    startingDataFrame = startingDataFrame.groupby(['date_range_start', 'state', 'state_code',\
                           'cnamelong', 'county_code']).agg(device_count = ('device_count', 'sum'),\
                                                            completely_home_device_count = ('completely_home_device_count', 'sum')).reset_index()
```

```python
startingDataFrame
```

startingDataFrame, which is generated by processing by chunks of data, matches our processing by entire dataset of this 50000 rows. Now we can proceed to apply this group by aggregate to our entire dataset.

```python
aggregatedData = pd.read_csv("social_dist_all_reduced.csv", nrows=0)
```

```python
counter = 0
for chunk in pd.read_csv("social_dist_all_reduced.csv", chunksize = 1000000):
    chunk['device_count'] = pd.to_numeric(chunk['device_count'], errors='coerce')
    chunk['completely_home_device_count'] = pd.to_numeric(chunk['completely_home_device_count'], errors='coerce')
    aggregatedData = aggregatedData.append(chunk, ignore_index = True)
    aggregatedData = aggregatedData.groupby(['date_range_start', 'state', 'state_code',\
                                             'cnamelong', 'county_code']).agg(device_count = ('device_count', 'sum'),\
                                                                              completely_home_device_count = ('completely_home_device_count', 'sum')).reset_index()
    print(counter) # track progress
    print(aggregatedData.shape[0]) # how large is our data right now
    counter += 1
```
