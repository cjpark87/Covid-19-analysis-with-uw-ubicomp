```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data_path = '/content/drive/MyDrive/UW/Ubicomp/COVID19 data analysis/Covid-19-analysis-with-uw-ubicomp/final_analysis/data/'
df_safegraph = pd.read_csv(data_path+'safegraph_data_by_county_diff_baseline.csv')
df_safegraph = df_safegraph[(df_safegraph['date_range_start'] >= '2020-02-17') & (df_safegraph['date_range_start'] <= '2020-12-13')]
df_safegraph.to_csv(data_path+'safegraph_data_by_county_diff_baseline.csv', index=False)
df_google = pd.read_csv(data_path+'google_mobility.csv')
df_google = df_google[(df_google['date'] >= '2020-02-17') & (df_google['date'] <= '2020-12-13')]
for c in df_google.columns:
  if 'Unnamed' in c:
    del df_google[c]
df_google.to_csv(data_path+'google_mobility.csv', index=False)

print(df_safegraph.columns)
print(df_google.columns)
```
