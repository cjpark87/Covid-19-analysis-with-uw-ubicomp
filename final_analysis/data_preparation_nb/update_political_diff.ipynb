```
import pandas as pd
```

```
df_election = pd.read_csv('2020_US_County_Level_Presidential_Results.csv')
df_election['political_party'] = df_election.apply(lambda x: 'Republican' if x['per_gop'] > x['per_dem'] else 'Democratic', axis=1)
df_election['political_diff'] = df_election.apply(lambda x: x['per_dem'] - x['per_gop'], axis=1)
```

```
df_election_2016 = pd.read_csv('2016_US_County_Level_Presidential_Results.csv')
df_covariates = pd.read_csv('county_data_with_covariates.csv')
print(df_covariates.columns)
print(df_election.columns)
for c in df_election_2016.columns:
  if c in df_covariates.columns:
    del df_covariates[c]  

for c in df_election.columns:
  if c in df_covariates.columns:
    del df_covariates[c]  

df_covariates = df_covariates.merge(df_election, left_on='county_code', right_on='county_fips')
df_covariates.to_csv('county_data_with_covariates_new.csv')
```
